Em primeiro lugar, encontramos a lista de S & P100 na internet. A fonte mais óbvia é a página da Wikipédia para empresas S&P100 (https://en.wikipedia.org/wiki/S%26P_100). Além disso, a Wikipédia tem uma maneira sistemática de criar suas páginas html que permite a fácil raspagem de dados.

Assim, obtemos a lista de todos os tickers no S&P100 usando o código abaixo:

```python
#extracting data from Yahoo Finance API
#Obtain list of S&100 companies from wikipedia
resp = requests.get("https://en.wikipedia.org/wiki/S%26P_100")
convert_soup = bs.BeautifulSoup(resp.text, 'lxml')
table = convert_soup.find('table',{'class':'wikitable sortable'})

tickers = []

for rows in table.findAll('tr')[1:]:
    ticker = rows.findAll('td')[0].text.strip()
    tickers.append(ticker)

all_data = pd.DataFrame()
test_data = pd.DataFrame()
no_data = []

for i in tickers:
    try:
        print(i)
        test_data = yf.download(i, interval = '1d', start = dt.datetime(1990,1,1), end = dt.date.today())
        test_data['symbol'] = i
        all_data = all_data.append(test_data)
        clear_output(wait = True)
    except:
        no_data.append(i)

    clear_output(wait = True)
```

### Indicadores

```python
all_data['SMA_5'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window = 5).mean())
all_data['SMA_15'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window = 15).mean())
all_data['SMA_ratio'] = all_data['SMA_15'] / all_data['SMA_5']

all_data['SMA5_Volume'] = all_data.groupby('symbol')['Volume'].transform(lambda x: x.rolling(window = 5).mean())
all_data['SMA15_Volume'] = all_data.groupby('symbol')['Volume'].transform(lambda x: x.rolling(window = 15).mean())
all_data['SMA_Volume_Ratio'] = all_data['SMA5_Volume']/all_data['SMA15_Volume']
```

```python
def Wilder(data, periods):
    start = np.where(~np.isnan(data))[0][0] #Check if nans present in beginning
    Wilder = np.array([np.nan]*len(data))
    Wilder[start+periods-1] = data[start:(start+periods)].mean() #Simple Moving Average
    for i in range(start+periods,len(data)):
        Wilder[i] = (Wilder[i-1]*(periods-1) + data[i])/periods #Wilder Smoothing
    return(Wilder)

all_data['prev_close'] = all_data.groupby('symbol')['Close'].shift(1)
all_data['TR'] = np.maximum((all_data['High'] - all_data['Low']), 
                     np.maximum(abs(all_data['High'] - all_data['prev_close']), 
                     abs(all_data['prev_close'] - all_data['Low'])))
for i in all_data['symbol'].unique():
    TR_data = all_data[all_data.symbol == i].copy()
    all_data.loc[all_data.symbol==i,'ATR_5'] = Wilder(TR_data['TR'], 5)
    all_data.loc[all_data.symbol==i,'ATR_15'] = Wilder(TR_data['TR'], 15)

all_data['ATR_Ratio'] = all_data['ATR_5'] / all_data['ATR_15']


all_data['prev_high'] = all_data.groupby('symbol')['High'].shift(1)
all_data['prev_low'] = all_data.groupby('symbol')['Low'].shift(1)

all_data['+DM'] = np.where(~np.isnan(all_data.prev_high),
                           np.where((all_data['High'] > all_data['prev_high']) & 
         (((all_data['High'] - all_data['prev_high']) > (all_data['prev_low'] - all_data['Low']))), 
                                                                  all_data['High'] - all_data['prev_high'], 
                                                                  0),np.nan)

all_data['-DM'] = np.where(~np.isnan(all_data.prev_low),
                           np.where((all_data['prev_low'] > all_data['Low']) & 
         (((all_data['prev_low'] - all_data['Low']) > (all_data['High'] - all_data['prev_high']))), 
                                    all_data['prev_low'] - all_data['Low'], 
                                    0),np.nan)

for i in all_data['symbol'].unique():
    ADX_data = all_data[all_data.symbol == i].copy()
    all_data.loc[all_data.symbol==i,'+DM_5'] = Wilder(ADX_data['+DM'], 5)
    all_data.loc[all_data.symbol==i,'-DM_5'] = Wilder(ADX_data['-DM'], 5)
    all_data.loc[all_data.symbol==i,'+DM_15'] = Wilder(ADX_data['+DM'], 15)
    all_data.loc[all_data.symbol==i,'-DM_15'] = Wilder(ADX_data['-DM'], 15)

all_data['+DI_5'] = (all_data['+DM_5']/all_data['ATR_5'])*100
all_data['-DI_5'] = (all_data['-DM_5']/all_data['ATR_5'])*100
all_data['+DI_15'] = (all_data['+DM_15']/all_data['ATR_15'])*100
all_data['-DI_15'] = (all_data['-DM_15']/all_data['ATR_15'])*100

all_data['DX_5'] = (np.round(abs(all_data['+DI_5'] - all_data['-DI_5'])/(all_data['+DI_5'] + all_data['-DI_5']) * 100))

all_data['DX_15'] = (np.round(abs(all_data['+DI_15'] - all_data['-DI_15'])/(all_data['+DI_15'] + all_data['-DI_15']) * 100))

for i in all_data['symbol'].unique():
    ADX_data = all_data[all_data.symbol == i].copy()
    all_data.loc[all_data.symbol==i,'ADX_5'] = Wilder(ADX_data['DX_5'], 5)
    all_data.loc[all_data.symbol==i,'ADX_15'] = Wilder(ADX_data['DX_15'], 15)

all_data['Lowest_5D'] = all_data.groupby('symbol')['Low'].transform(lambda x: x.rolling(window = 5).min())
all_data['High_5D'] = all_data.groupby('symbol')['High'].transform(lambda x: x.rolling(window = 5).max())
all_data['Lowest_15D'] = all_data.groupby('symbol')['Low'].transform(lambda x: x.rolling(window = 15).min())
all_data['High_15D'] = all_data.groupby('symbol')['High'].transform(lambda x: x.rolling(window = 15).max())

all_data['Stochastic_5'] = ((all_data['Close'] - all_data['Lowest_5D'])/(all_data['High_5D'] - all_data['Lowest_5D']))*100
all_data['Stochastic_15'] = ((all_data['Close'] - all_data['Lowest_15D'])/(all_data['High_15D'] - all_data['Lowest_15D']))*100

all_data['Stochastic_%D_5'] = all_data['Stochastic_5'].rolling(window = 5).mean()
all_data['Stochastic_%D_15'] = all_data['Stochastic_5'].rolling(window = 15).mean()

all_data['Stochastic_Ratio'] = all_data['Stochastic_%D_5']/all_data['Stochastic_%D_15']

all_data['Diff'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.diff())
all_data['Up'] = all_data['Diff']
all_data.loc[(all_data['Up']<0), 'Up'] = 0

all_data['Down'] = all_data['Diff']
all_data.loc[(all_data['Down']>0), 'Down'] = 0 
all_data['Down'] = abs(all_data['Down'])

all_data['avg_5up'] = all_data.groupby('symbol')['Up'].transform(lambda x: x.rolling(window=5).mean())
all_data['avg_5down'] = all_data.groupby('symbol')['Down'].transform(lambda x: x.rolling(window=5).mean())

all_data['avg_15up'] = all_data.groupby('symbol')['Up'].transform(lambda x: x.rolling(window=15).mean())
all_data['avg_15down'] = all_data.groupby('symbol')['Down'].transform(lambda x: x.rolling(window=15).mean())

all_data['RS_5'] = all_data['avg_5up'] / all_data['avg_5down']
all_data['RS_15'] = all_data['avg_15up'] / all_data['avg_15down']

all_data['RSI_5'] = 100 - (100/(1+all_data['RS_5']))
all_data['RSI_15'] = 100 - (100/(1+all_data['RS_15']))

all_data['RSI_ratio'] = all_data['RSI_5']/all_data['RSI_15']

all_data['5Ewm'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.ewm(span=5, adjust=False).mean())
all_data['15Ewm'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.ewm(span=15, adjust=False).mean())
all_data['MACD'] = all_data['15Ewm'] - all_data['5Ewm']

all_data['15MA'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window=15).mean())
all_data['SD'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window=15).std())
all_data['upperband'] = all_data['15MA'] + 2*all_data['SD']
all_data['lowerband'] = all_data['15MA'] - 2*all_data['SD']

all_data['RC'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.pct_change(periods = 15)) 

all_data['Close_Shifted'] = all_data.groupby('symbol')['Close'].transform(lambda x: x.shift(-6))
all_data['Target'] = ((all_data['Close_Shifted'] - all_data['Open'])/(all_data['Open']) * 100).shift(-1)
all_data['Target_Direction'] = np.where(all_data['Target']>0,1,0)
all_data = all_data.dropna().copy()
```

### **Winsorizando os indicadores**

   Antes de começar a desenvolver o modelo de previsão, é importante lidar com os valores atípicos que existem em nas variáveis explicativas, ou seja, nossos Indicadores Técnicos. Existem várias maneiras pelas quais os valores atípicos podem ser tratados, uma das quais é a winsorizing dos dados. A ideia de winsorizing é trazer outliers extremos para o valor mais próximo que não é considerado um outlier. Por exemplo, aqui nós winsorize 10% menores e 10% mais altos valores, assim, os valores extremos agora mudarão para os valores do percentil 10 e do percentil 90, respectivamente. Uma das principais vantagens do winsorizing é que as informações contidas nos outliers extremos não são perdidas; apenas os valores absolutos desses são sensibilizados.

```python
Target_variables = ['SMA_ratio','ATR_5','ATR_15','ATR_Ratio',
                       'ADX_5','ADX_15','SMA_Volume_Ratio','Stochastic_5','Stochastic_15','Stochastic_Ratio',
                      'RSI_5','RSI_15','RSI_ratio','MACD']
for variable in Target_variables:
    all_data.loc[:,variable] = mstats.winsorize(all_data.loc[:,variable], limits = [0.1,0.1])

all_data['return'] = all_data.groupby('symbol')['Close'].pct_change() 
```

### **Empresas de agrupamento**

É aqui que as coisas começam a ficar interessantes. Uma maneira simples de prever seria assumir que todas as empresas seguiriam o mesmo modelo de ML e criariam esse modelo global para prever retornos para todas as empresas. No entanto, é possível que diferentes empresas/indústrias reajam de forma diferente a um conjunto de Indicadores Técnicos. Uma maneira de resolver esse problema é criar um modelo de ML diferente para cada cluster de empresas que se espera que se comportem de forma semelhante, talvez pertencentes ao mesmo setor, onde o "comportamento" é capturado em seus retornos.

Assim, para tornar o modelo ainda mais sofisticado, serão criados diferentes modelos de ML para cada cluster.

**OBS: Pesquisar Teoria**


**Identificar o número de clusters**

A ideia é usar retornos dessas empresas e criar uma curva para determinar o número de clusters que criariam um equilíbrio entre a baixa soma de quadrados dentro de um cluster versus o número total de clusters. Para 100 empresas, se criássemos 100 clusters, teríamos 0 soma de quadrados dentro do cluster, no entanto, esse clustering não faria sentido, ou seja, não seria parcimonioso. Por outro lado, ter 1 cluster seria parcimonioso, no entanto, levará a uma soma muito alta de quadrados dentro do cluster. Uma curva de cotovelo ajuda a determinar o ponto aproximado em que a diminuição marginal na soma dos quadrados é pequena.

Usa-se o agrupamento K-means para criar uma curva. K-means visa minimizar a inércia ou a soma dentro do cluster de quadrados durante o agrupamento. Ao fornecer uma gama de clusters de 1 a 50, criamos a curva de cotovelo necessária.

```python
#Extract the returns
returns = all_data[['symbol','return']].copy()
returns['Date'] = returns.index.copy()

#Pivot the returns to create series of returns for each stock 
transposed = returns.pivot(index = 'Date', columns = 'symbol', values = 'return')

#Transpose the data to get companies on the index level and dates on the column level since clusters takes place on index level
X = transposed.dropna().transpose()

#Extract sum of squares for K-means clusters from 1 to 50 clusters
sum_of_sq = np.zeros([50, 1])
for k in range(1, 51):
    sum_of_sq[k-1] = KMeans(n_clusters=k).fit(X).inertia_
    
plt.plot(range(1, 50), sum_of_sq[1:50])
plt.title("Elbow Method") 
plt.xlabel("Number of Cluster") 
plt.ylabel("Within-cluster Sum of Squares")

pd.DataFrame(sum_of_sq, columns = ['Difference in SS'], index = range(1,51)).diff()
```

### **Modelo de Random Forest**

Random Forest é um modelo de Machine Learning comumente usado para problemas de Regressão e Classificação. No entanto, dada a complexidade do modelo, é importante entender cuidadosamente os parâmetros que entram no modelo para evitar o sobreajuste ou o subajuste na amostra, uma troca padrão de viés-variância.

Há algumas coisas a considerar ao formar um modelo de Machine Learning. Vamos passar um por um.

**Período de formação**

Primeiro, separaremos nossos dados em um treino e uma amostra de teste. **Tipo de modelo de Random Forest**

A Floresta Aleatória tem principalmente dois tipos — Regressor e Classificador. Regressor é quando queremos predição em uma forma de regressão não-linear, enquanto Classifier é usado como um modelo logístico não-linear. Usaremos o Classificador como queremos avaliar a probabilidade de um movimento para cima (probabilidade de 1) para cada ação. Semelhante à regressão logística, o Random Forest Classifier nos dará exatamente isso.

**Ajuste de hiperparâmetros**

Uma das principais etapas de um modelo de floresta aleatória é ajustar os parâmetros da floresta aleatória. Random Forest tem de 5 a 6 parâmetros-chave que definem como a floresta será construída. São eles: max_depth, max_features, min_samples_leaf, n_estimators min_samples_split. Como esses parâmetros podem ter uma ampla gama de valores, usaremos um GridSearchCV para obter a melhor combinação desses parâmetros para cada cluster.

O GridSearchCV trabalha com as combinações possíveis desses valores de parâmetros que fornecemos e fornece a melhor combinação que teria o menor erro na validação cruzada fora da amostra. Esses parâmetros serão então usados para previsão.

**Curvas de validação**

Para determinar os valores iniciais que serão dados ao GridSearchCV, sobre os quais ele trabalhará para encontrar a melhor combinação, podemos usar Curvas de Validação para cada um dos parâmetros. As curvas de validação também analisam a validação cruzada e fornecem uma pontuação de previsão para dentro e fora da amostra. Isso nos fornece uma boa ideia do valor inicial em torno do qual podemos fornecer um intervalo para o GridSearchCV.

Abaixo está um exemplo de pontuação da curva de validação para n_estimators. No n_estimators mais baixo, vemos o underfitting, enquanto no n_estimators mais alto vemos o over-fitting. O underfitting leva a uma variância de alto viés-baixa, onde o overfitting leva a uma variância baixa de viés-alta. Podemos ver que em n=7, obtemos um bom tradeoff entre viés e variância. Portanto, em nosso GridSearchCV, usamos n_estimators para ser em torno de 7.

```python
#Separate between X and Y
X_train = train_data.loc[:,Target_variables]

Y_train = train_data.loc[:,['Target_Direction']]

#Create validation curve for the Random Forest Classifier
rf = RandomForestClassifier()
train_scoreNum, test_scoreNum = validation_curve(rf,
                                X = X_train['2010-01-01':], y = Y_train.loc['2010-01-01':,'Target_Direction'], 
                                param_name = 'n_estimators', 
                                param_range = [3,4,7,10,12,15,20,25,30], cv = TimeSeriesSplit(n_splits = 3))

train_scores_mean = np.mean(train_scoreNum, axis=1)
train_scores_std = np.std(train_scoreNum, axis=1)
test_scores_mean = np.mean(test_scoreNum, axis=1)
test_scores_std = np.std(test_scoreNum, axis=1)

plt.figure(figsize = (20,10))
plt.plot([3,4,7,10,12,15,20,25,30],train_scores_mean)
plt.plot([3,4,7,10,12,15,20,25,30],test_scores_mean)
plt.legend(['Train Score','Test Score'], fontsize = 'large')
plt.title('Validation Curve Score for n_estimators', fontsize = 'large')
```

**Pickle files**

Python oferece uma maneira muito conveniente de salvar arquivos de função usando o pacote pickle. A ideia seria ajustar o modelo nos dados e salvar esse modelo ajustado em arquivos de picles para cada cluster. Dessa forma, teríamos um modelo diferente salvo para cada cluster. Assim, ao prever para uma determinada empresa, usaremos o modelo no arquivo de picles do cluster correspondente e faremos nossa previsão.

**OBS: Pesquisar Teoria**

**Random Forest Model**

```python
#Run the loop for every unique cluster - 17 loops
for cluster_selected in clusters_df.Cluster.unique():
    
    print(f'The current cluster running is : {cluster_selected}')
    
    #Get data for that cluster
    co_data = all_data[all_data.symbol.isin(clusters_df.loc[clusters_df.Cluster==cluster_selected,'Companies'].tolist())].copy()
    co_train = co_data[:'2018-12-31']
    co_train = co_train.dropna().copy()
    
    X_train = co_train.loc[:,Target_variables]

    Y_train = co_train.loc[:,['Target_Direction']]

    #Define paramters from Validation Curve
    params = {'max_depth': [5, 7],
          'max_features': ['sqrt'],
          'min_samples_leaf': [10, 15, 20],
          'n_estimators': [5, 7, 9],
         'min_samples_split':[20, 25, 30]} #Using Validation Curves

    rf = RandomForestClassifier()

    #Perform a TimeSeriesSplit on the dataset
    time_series_split = TimeSeriesSplit(n_splits = 3)

    
    rf_cv = GridSearchCV(rf, params, cv = time_series_split, n_jobs = -1, verbose = 20)

    #Fit the random forest with our X_train and Y_train
    rf_cv.fit(X_train, Y_train)
          
    #Save the fited variable into a Pickle file
    file_loc = f'{os.getcwd()}\\Pickle_Files\\Cluster_{cluster_selected}'    
    pickle.dump(rf_cv, open(file_loc,'wb'))
```

```python
# Simple Moving Average (SMA)

def SMA(data):
    data['SMA_5'] = data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window = 5).mean())
    data['SMA_15'] = data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window = 15).mean())
    data['SMA_ratio'] = data['SMA_15'] / data['SMA_5']

    return data

# Simple Moving Average Volume

def SMAV(data):
    data['SMA5_Volume'] = data.groupby('symbol')['Volume'].transform(lambda x: x.rolling(window = 5).mean())
    data['SMA15_Volume'] = data.groupby('symbol')['Volume'].transform(lambda x: x.rolling(window = 15).mean())
    data['SMA_Volume_Ratio'] = data['SMA5_Volume']/data['SMA15_Volume']

    return data
```

```python
# Wilder’s Smoothing
def Wilder(data, periods):
    start = np.where(~np.isnan(data))[0][0] #Check if nans present in beginning
    Wilder = np.array([np.nan]*len(data))
    Wilder[start+periods-1] = data[start:(start+periods)].mean() #Simple Moving Average
    for i in range(start+periods,len(data)):
        Wilder[i] = (Wilder[i-1]*(periods-1) + data[i])/periods #Wilder Smoothing
    return (Wilder)

# Average True Range (ATR)
def ATR(data):
    data['prev_close'] = data.groupby('symbol')['Close'].shift(1)
    data['TR'] = np.maximum((data['High'] - data['Low']), 
                        np.maximum(abs(data['High'] - data['prev_close']), 
                        abs(data['prev_close'] - data['Low'])))
    for i in data['symbol'].unique():
        TR_data = data[data.symbol == i].copy()
        data.loc[data.symbol==i,'ATR_5'] = Wilder(TR_data['TR'], 5)
        data.loc[data.symbol==i,'ATR_15'] = Wilder(TR_data['TR'], 15)

    data['ATR_Ratio'] = data['ATR_5'] / data['ATR_15']

    return data

# Average Directional Index (ADX)
def ADX(data):
    data['prev_high'] = data.groupby('symbol')['High'].shift(1)
    data['prev_low'] = data.groupby('symbol')['Low'].shift(1)

    data['+DM'] = np.where(~np.isnan(data.prev_high),
                              np.where((data['High'] > data['prev_high']) & 
            (((data['High'] - data['prev_high']) > (data['prev_low'] - data['Low']))), 
                                                                      data['High'] - data['prev_high'], 
                                                                      0),np.nan)

    data['-DM'] = np.where(~np.isnan(data.prev_low),
                              np.where((data['prev_low'] > data['Low']) & 
            (((data['prev_low'] - data['Low']) > (data['High'] - data['prev_high']))), 
                                        data['prev_low'] - data['Low'], 
                                        0),np.nan)

    for i in data['symbol'].unique():
        ADX_data = data[data.symbol == i].copy()
        data.loc[data.symbol==i,'+DM_5'] = Wilder(ADX_data['+DM'], 5)
        data.loc[data.symbol==i,'-DM_5'] = Wilder(ADX_data['-DM'], 5)
        data.loc[data.symbol==i,'+DM_15'] = Wilder(ADX_data['+DM'], 15)
        data.loc[data.symbol==i,'-DM_15'] = Wilder(ADX_data['-DM'], 15)

    data['+DI_5'] = (data['+DM_5']/data['ATR_5'])*100
    data['-DI_5'] = (data['-DM_5']/data['ATR_5'])*100
    data['+DI_15'] = (data['+DM_15']/data['ATR_15'])*100
    data['-DI_15'] = (data['-DM_15']/data['ATR_15'])*100

    data['DX_5'] = (np.round(abs(data['+DI_5'] - data['-DI_5'])/(data['+DI_5'] + data['-DI_5']) * 100))

    data['DX_15'] = (np.round(abs(data['+DI_15'] - data['-DI_15'])/(data['+DI_15'] + data['-DI_15']) * 100))

    for i in data['symbol'].unique():
        ADX_data = data[data.symbol == i].copy()
        data.loc[data.symbol==i,'ADX_5'] = Wilder(ADX_data['DX_5'], 5)
        data.loc[data.symbol==i,'ADX_15'] = Wilder(ADX_data['DX_15'], 15)

    return data

# Stochastic Oscillators
def Stoc(data):
    data['Lowest_5D'] = data.groupby('symbol')['Low'].transform(lambda x: x.rolling(window = 5).min())
    data['High_5D'] = data.groupby('symbol')['High'].transform(lambda x: x.rolling(window = 5).max())
    data['Lowest_15D'] = data.groupby('symbol')['Low'].transform(lambda x: x.rolling(window = 15).min())
    data['High_15D'] = data.groupby('symbol')['High'].transform(lambda x: x.rolling(window = 15).max())

    data['Stochastic_5'] = ((data['Close'] - data['Lowest_5D'])/(data['High_5D'] - data['Lowest_5D']))*100
    data['Stochastic_15'] = ((data['Close'] - data['Lowest_15D'])/(data['High_15D'] - data['Lowest_15D']))*100

    data['Stochastic_%D_5'] = data['Stochastic_5'].rolling(window = 5).mean()
    data['Stochastic_%D_15'] = data['Stochastic_5'].rolling(window = 15).mean()

    data['Stochastic_Ratio'] = data['Stochastic_%D_5']/data['Stochastic_%D_15']

    return data

# Relative Strength Index (RSI)
def RSI(data):
    data['Diff'] = data.groupby('symbol')['Close'].transform(lambda x: x.diff())
    data['Up'] = data['Diff']
    data.loc[(data['Up']<0), 'Up'] = 0

    data['Down'] = data['Diff']
    data.loc[(data['Down']>0), 'Down'] = 0 
    data['Down'] = abs(data['Down'])

    data['avg_5up'] = data.groupby('symbol')['Up'].transform(lambda x: x.rolling(window=5).mean())
    data['avg_5down'] = data.groupby('symbol')['Down'].transform(lambda x: x.rolling(window=5).mean())

    data['avg_15up'] = data.groupby('symbol')['Up'].transform(lambda x: x.rolling(window=15).mean())
    data['avg_15down'] = data.groupby('symbol')['Down'].transform(lambda x: x.rolling(window=15).mean())

    data['RS_5'] = data['avg_5up'] / data['avg_5down']
    data['RS_15'] = data['avg_15up'] / data['avg_15down']

    data['RSI_5'] = 100 - (100/(1+data['RS_5']))
    data['RSI_15'] = 100 - (100/(1+data['RS_15']))

    data['RSI_ratio'] = data['RSI_5']/data['RSI_15']

    return data


# Moving Average Convergence Divergence (MACD)
def MACD(data):
    data['5Ewm'] = data.groupby('symbol')['Close'].transform(lambda x: x.ewm(span=5, adjust=False).mean())
    data['15Ewm'] = data.groupby('symbol')['Close'].transform(lambda x: x.ewm(span=15, adjust=False).mean())
    data['MACD'] = data['15Ewm'] - data['5Ewm']

    return data

# Bollinger Bands
def Bol(data):
    data['15MA'] = data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window=15).mean())
    data['SD'] = data.groupby('symbol')['Close'].transform(lambda x: x.rolling(window=15).std())
    data['upperband'] = data['15MA'] + 2*data['SD']
    data['lowerband'] = data['15MA'] - 2*data['SD']

    return data

# Rate of Change
def RC(data):
    data['RC'] = data.groupby('symbol')['Close'].transform(lambda x: x.pct_change(periods = 15)) 

    data['Close_Shifted'] = data.groupby('symbol')['Close'].transform(lambda x: x.shift(-6))
    data['Target'] = ((data['Close_Shifted'] - data['Open'])/(data['Open']) * 100).shift(-1)
    data['Target_Direction'] = np.where(data['Target']>0,1,0)
    data = data.dropna().copy()

    return data
```

```python
def n_cluster(data):
    #Extract the returns
    returns = data[['symbol','return']].copy()
    returns['Date'] = returns.index.copy()

    #Pivot the returns to create series of returns for each stock 
    transposed = returns.pivot(index = 'Date', columns = 'symbol', values = 'return')

    #Transpose the data to get companies on the index level and dates on the column level since clusters takes place on index level
    X = transposed.dropna().transpose()

    #Extract sum of squares for K-means clusters from 1 to 50 clusters
    sum_of_sq = np.zeros([50, 1])
    for k in range(1, 51):
        sum_of_sq[k-1] = KMeans(n_clusters=k).fit(X).inertia_

    return sum_of_sq, transposed
        
def plot_cluster(sum_of_sq):

    plt.plot(range(1, 50), sum_of_sq[1:50])
    plt.title("Elbow Method") 
    plt.xlabel("Number of Cluster") 
    plt.ylabel("Within-cluster Sum of Squares")

    print(pd.DataFrame(sum_of_sq, columns = ['Difference in SS'], index = range(1,51)).diff())
```

```python
def cluster(transposed):
    n = int(input(" Escolha o número de clusters desejado com base nas opções apresentadas: "))

    #Get n clusters
    gmm = GaussianMixture(n_components = n)
    gmm.fit(transposed.dropna().transpose())

    #Predict for each company
    clusters = gmm.predict(transposed.dropna().transpose())
    clusters_df = pd.DataFrame({'Cluster':clusters,
                              'Companies':transposed.columns})

    #Sort by Clusters
    clusters_df = clusters_df.sort_values(['Cluster']).reset_index(drop = True)

    #Save as csv
    clusters_df.to_csv("clusters.csv")
    clusters_df = pd.read_csv("clusters.csv", index_col = 0)

    return clusters_df
```
